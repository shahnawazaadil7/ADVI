t-SNE (t-distributed stochastic neighbor embedding) is a statistical method that visualizes high-dimensional data by placing each data point in a two or three-dimensional map

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(n_samples=1000, centers=3, n_features=2, random_state=0)
tsne = TSNE(n_components=2, perplexity = 80, random_state=42)
```

```python
x_reduced = tsne.fit_transform(x)
```

```python
plt.scatter(x_reduced[:,0],x_reduced[:,1], c =y)
plt.title('t-SNE')
```

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()
X, y = digits.data, digits.target

perplexities = [5, 30, 50]
fig, axes = plt.subplots(1, len(perplexities), figsize=(15, 5))

for i, perplexity in enumerate(perplexities):
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    X_embedded = tsne.fit_transform(X)
    
    axes[i].scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap='Spectral', s=10)
    axes[i].set_title(f"Perplexity: {perplexity}")
    axes[i].axis('off')

plt.show()
```
