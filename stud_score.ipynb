```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
np.random.seed(0)
hours_studied = np.random.uniform(0, 10, 100)
exam_scores = 5 * hours_studied + np.random.normal(0, 5, 100)
```

```python
plt.hist(exam_scores, bins=20)
```

```python
#hours_studied = (hours_studied - np.mean(hours_studied)) / np.std(hours_studied)
#exam_scores = (exam_scores - np.mean(exam_scores)) / np.std(exam_scores)
```

```python
X = np.vstack((np.ones(hours_studied.size), hours_studied)).T
y = exam_scores
```

```python
theta = np.linalg.inv(X.T @ X) @ X.T @ y
```

```python
intercept, slope = theta
```

```python
predicted_scores = X @ theta
```

```python
plt.scatter(hours_studied, exam_scores, color='blue', label='Actual Scores')

plt.plot(hours_studied, predicted_scores, color='red', label='Predicted Scores')

plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Student Exam Scores Prediction')
plt.legend()

plt.show()
```

```python
def predict_marks(hours):
    return intercept + slope * hours
```

```python
predict_marks(12)
```
