# Feature Engineering
Feature scaling is a technique used in machine learning to normalize or standardize the range of independent variables (features) in a dataset.

The techniques involved in the feature engineering is encoding categorical data, dimensionality reduction 


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/shahnawazaadil/Downloads/Social_Network_Ads.csv')
```

```python
df
```

```python
df = df.iloc[:,2:]
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('Purchased',axis=1), df['Purchased'],test_size=0.3,random_state=0)
```

```python
x_train.shape
```

```python
y_train
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
scaler.mean_
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns=x_train.columns)
```

```python
x_test_scaled = pd.DataFrame(x_test_scaled, columns = x_test.columns)
```

```python
x_train_scaled
```

```python
np.round(x_train_scaled.describe())
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.scatter(x_train['Age'], x_train['EstimatedSalary'])
ax1.set_title('Before Scaling')
ax2.scatter(x_train_scaled['Age'],x_train_scaled['EstimatedSalary'],color='red')
ax2.set_title('After Scaling')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Age'],ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'],ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(x_train_scaled['Age'], ax=ax2)
sns.kdeplot(x_train_scaled['EstimatedSalary'],ax=ax2)
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Age'], ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(x_train_scaled['Age'], ax=ax2, color='orange')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, figsize=(12,5))

ax1.set_title('Before Scaling')
sns.kdeplot(x_train['EstimatedSalary'], ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(x_train_scaled['EstimatedSalary'], ax=ax2, color='orange')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_scaled.fit(x_train_scaled,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred_scaled=lr_scaled.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred_scaled)
```

```python
wine = pd.read_csv('/Users/shahnawazaadil/Downloads/wine_data.csv', header=None,usecols=[0,1,2])
```

```python
wine.columns=['Class Labels', 'Alcohol', 'Malic Acid']
```

```python
color_dict={1:'red',2:'green',3:'blue'}
```

```python
sns.scatterplot(x=wine['Alcohol'],y=wine['Malic Acid'],hue=wine['Class Labels'],palette=color_dict)
```

```python
x_train,x_test,y_train,y_test = train_test_split(wine.drop('Class Labels', axis=1), 
                                                 wine['Class Labels'],test_size=0.3,random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_scaled=scaler.transform(x_train)
```

```python
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
```

```python
x_test_scaled= pd.DataFrame(x_test_scaled, columns=x_test.columns)
```

```python
np.round(x_train.describe(),1)
```

```python
np.round(x_train_scaled.describe(),1)
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2, figsize=(12,5))
ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Alcohol'],ax=ax1)
sns.kdeplot(x_train['Malic Acid'],ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(x_train_scaled['Alcohol'], ax=ax2)
sns.kdeplot(x_train_scaled['Malic Acid'], ax=ax2)
plt.show()
```
